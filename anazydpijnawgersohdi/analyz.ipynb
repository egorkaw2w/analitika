Импортируем pandas, загружаем датасет "качество воды", в который входит информация о веществах, которые входят в состав воды из различных образцов ( алюминий, аммиак, мышьяк, барий, кадмий, хлорамин, хром, фторид меди, бактерии, нитраты свинца, нитриты, перхлорат ртути, радий, селен, серебро, уран.), а также, безопасна вода или нет

```python
import pandas as pd
per = pd.read_csv("waterQuality1.csv")
#грузим воду
per
```

 С помощью  shape  мы получаем количество строк и столбцов в датасете. В данном датасете 7999 строк и 21 столбец.

```python
per.shape
#получаем размерность вады
```

Здесь мы выводим названя столбцов

```python
print(per.columns)
#получаем названия вады
```

Здесь выводим уникальные значения из столбцов 

```python
for column in per.columns:
    print(per[column].unique())
    per
    #понимаем, что вода у нас не красивая
```

Здесь мы сортируем значения в столбце aluminium

```python
per = per.sort_values("aluminium")
per
#сортировка воды
```

Здесь мы удаляем дубликаты значений 

```python
per = per.drop_duplicates()
per
#удаляем дубликаты ^_^
```

Здесь мы заменяем все значения 0.00 в столбце aluminium на значение 1

```python
per["aluminium"] = per["aluminium"].replace(0.00, 1)
per
```

Здесь мы удаляем столбец ammonia

```python
per = per.drop("ammonia",axis=1)
per
```

Получаем общую информацию о датасете, количество непустых значений, типы данных в каждом столбце и объём потребляемой памяти 

```python
per.info()
```

Получаем статическое описание числовых данных датасета
